```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('Pokemontree.csv')
df
```

```python
%matplotlib inline
plt.xlabel('hp')
plt.ylabel('cp')
plt.scatter(df.HP,df.CP,color='red',marker='.')
```

```python
HP = df[['HP']]
HP
```

```python
CP = df.CP
CP
```

```python
reg = linear_model.LinearRegression()
reg.fit(HP,CP)
```

```python
hp_df = pd.read_csv("hp.csv")
hp_df.head(4)
```

```python
p = reg.predict(hp_df)
p
```

```python
hp_df['CP'] = p
hp_df
```

```python
%matplotlib inline
plt.xlabel('hp', fontsize=20)
plt.ylabel('cp', fontsize=20)
plt.scatter(df.HP,df.CP,color='red',marker='+')
plt.plot(df.HP,reg.predict(HP),color='blue')
```
